```python
# =========================================================
# Section 1 ─ 依赖导入 & 随机数种子（全 PyTorch 环境）
# =========================================================
import os, sys, re, time, itertools, random
os.environ["TORCH_DISABLE_DYNAMO"] = "1"
# export TORCH_COMPILE=0          # 完全关闭 torch.compile 及其依赖
# export TORCHDYNAMO_DISABLE=1    # 明确要求禁用 dynamo
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.utils import shuffle
from sklearn.pipeline import make_pipeline
from sklearn.decomposition import PCA
from scipy.spatial.distance import jensenshannon  # 保留此函数无需 TensorFlow

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from skopt import BayesSearchCV
from skopt.space import Real, Categorical, Integer
import joblib

# 设置全局随机数种子
SEED = 40
np.random.seed(SEED)
random.seed(SEED)
torch.manual_seed(SEED)

# =========================================================
# Section 2 ─ 评估函数 & JS Divergence 工具函数
# =========================================================
def calculate_accuracy_within_threshold(y_true, y_pred, threshold: float) -> float:
    """计算相对误差低于 threshold 的样本所占百分比。"""
    distances = [abs(r - p) / r if r != 0 else 0 for r, p in zip(y_true, y_pred)]
    return sum(d <= threshold for d in distances) / len(distances) * 100


def calculate_accuracy_cdf(y_true, y_pred, max_threshold: float = 1.0, step: float = 0.1):
    """返回 CDF 映射：{阈值: 精度百分比}。"""
    thresholds = np.arange(0, max_threshold + step, step)
    return {th: calculate_accuracy_within_threshold(y_true, y_pred, th) for th in thresholds}


def js_divergence(p: np.ndarray, q: np.ndarray) -> float:
    """JS Divergence（平方形式，与 TensorFlow 兼容）。"""
    return jensenshannon(p, q) ** 2


def calculate_js_divergence_between_datasets(df1: pd.DataFrame,
                                             df2: pd.DataFrame,
                                             features: list[str],
                                             bins: int = 30) -> dict[str, float]:
    """对若干特征计算两个数据集的 JS Divergence。"""
    divergences = {}
    for feat in features:
        if feat in df1.columns and feat in df2.columns:
            v_min = min(df1[feat].min(), df2[feat].min())
            v_max = max(df1[feat].max(), df2[feat].max())
            p_hist, _ = np.histogram(df1[feat], bins=bins, range=(v_min, v_max), density=True)
            q_hist, _ = np.histogram(df2[feat], bins=bins, range=(v_min, v_max), density=True)
            divergences[feat] = js_divergence(p_hist, q_hist)
    return divergences

# =========================================================
# Section 3 ─ 数据加载、CPU特征筛选与 JS Divergence 计算
# =========================================================
# 数据路径（按你的原始结构）
print("当前工作路径:", os.getcwd())

path1 = 'datasets/dag-1/resource_stimulus_global_A-B-C.csv'
path2 = 'datasets/dag-2/resource_stimulus_global_A-B-C.csv'
path3 = 'datasets/linear/resource_stimulus_global_A-B-C_modified.csv'
path4 = 'datasets/total/resource_stimulus_global_A-B-C_modified.csv'

print("数据路径设置完毕，准备加载")

replace_vals = {'<not counted>': np.nan, ' ': np.nan}

data_dag1   = pd.read_csv(path1, index_col=0).replace(replace_vals).dropna()
data_dag2   = pd.read_csv(path2, index_col=0).replace(replace_vals).dropna()
data_linear = pd.read_csv(path3, index_col=0).replace(replace_vals).dropna()
data_bridge = pd.read_csv(path4, index_col=0).replace(replace_vals).dropna()

print("数据加载完成")

# 筛选四个数据集中共同的特征列（排除 output_rate）
common_columns = (
    data_dag1.columns.intersection(data_dag2.columns)
             .intersection(data_linear.columns)
             .intersection(data_bridge.columns)
)
common_columns = common_columns[common_columns != 'output_rate']

# 筛选 CPU 相关特征
cpu_related_features = [
    c for c in data_bridge.columns
    if any(k in c for k in ('instructions', 'cycles', 'cache', 'branches'))
       and c not in ('input_rate', 'output_rate', 'latency')
]

# JS 散度计算
js_divergences = calculate_js_divergence_between_datasets(
    data_dag1, data_dag2, cpu_related_features, bins=30
)
average_js_divergence = float(np.mean(list(js_divergences.values())))

print(f"平均 JS Divergence = {average_js_divergence:.3f}")
```

```python
# =========================================================
# Section 4 ─ PCA & 数据准备
# =========================================================
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Bridge / Linear 作为训练集；Dag-1 / Dag-2 作为外部测试集
data_train1 = data_bridge[common_columns]
data_test1  = data_dag1[common_columns]

# ① 标准化
scaler = StandardScaler().fit(data_train1)
train1_scaled = scaler.transform(data_train1)
test1_scaled  = scaler.transform(data_test1)

# ② PCA：保留累计解释方差 ≥85%
pca_full = PCA().fit(train1_scaled)
n_components = np.where(np.cumsum(pca_full.explained_variance_ratio_) >= 0.85)[0][0] + 1
print(f"PCA n_components = {n_components}")

pca = PCA(n_components=n_components).fit(train1_scaled)
train1_pca = pca.transform(train1_scaled)
test1_pca  = pca.transform(test1_scaled)

cols_pca = [f"PC{i+1}" for i in range(n_components)]
df_train1 = pd.DataFrame(train1_pca, columns=cols_pca)
df_train1["output_rate"] = data_bridge["output_rate"].values
df_test1  = pd.DataFrame(test1_pca,  columns=cols_pca)
df_test1["output_rate"]  = data_dag1["output_rate"].values
```

```python
# =========================================================
# Section 5 ─ 基线 PyTorch MLP 训练 & 漂移检测
# =========================================================
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
from tqdm import tqdm

# 🔑 **y 不要 reshape，保持 1-D**
X = df_train1[cols_pca].values.astype(np.float32)
y = df_train1["output_rate"].values.astype(np.float32)       
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.3, random_state=SEED
)

class MLP(nn.Module):
    def __init__(self, in_dim, hidden1=32, hidden2=16):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_dim, hidden1),
            nn.ReLU(),
            nn.Linear(hidden1, hidden2),
            nn.ReLU(),
            nn.Linear(hidden2, 1),
        )
    def forward(self, x):
        return self.net(x).squeeze(1)   # 输出 [B]

def train_simple_mlp(X_tr, y_tr, X_val, y_val,
                     lr=1e-2, epochs=100, batch=8):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model  = MLP(X_tr.shape[1]).to(device)
    opt    = torch.optim.Adam(model.parameters(), lr=lr)
    loss_fn= nn.MSELoss()
    ds_tr  = TensorDataset(torch.from_numpy(X_tr), torch.from_numpy(y_tr))
    dl_tr  = DataLoader(ds_tr, batch_size=batch, shuffle=True)

    for epoch in range(epochs):
        model.train()
        with tqdm(dl_tr, desc=f"Epoch {epoch+1}/{epochs}", leave=False) as bar:
            for xb, yb in bar:
                xb, yb = xb.to(device), yb.to(device)
                pred = model(xb)                     # shape [B]
                loss = loss_fn(pred, yb)             # yb 也是 [B]
                opt.zero_grad(); loss.backward(); opt.step()
    # 验证
    model.eval()
    with torch.no_grad():
        y_hat = model(torch.from_numpy(X_val).to(device)).cpu().numpy()
    return model, y_hat

model_base, y_pred_val = train_simple_mlp(X_train, y_train, X_val, y_val)

thr = 0.15
acc_base = calculate_accuracy_within_threshold(y_val, y_pred_val, thr)
print(f"Baseline accuracy (val) = {acc_base:.2f}%")

# Dag-1
X_dag1 = df_test1[cols_pca].values.astype(np.float32)
y_dag1 = df_test1["output_rate"].values.astype(np.float32)

with torch.no_grad():
    y_pred_dag1 = model_base(torch.from_numpy(X_dag1)).numpy()
acc_dag1 = calculate_accuracy_within_threshold(y_dag1, y_pred_dag1, thr)
print(f"Accuracy on Dag-1 = {acc_dag1:.2f}%")
```

```python
# =========================================================
# Section 6 ─ skorch + GridSearchCV
# =========================================================
from skorch import NeuralNetRegressor
from skorch.callbacks import EarlyStopping
from sklearn.model_selection import GridSearchCV

net = NeuralNetRegressor(
    module           = MLP,
    module__in_dim   = X_train.shape[1],
    max_epochs       = 100,
    lr               = 0.01,
    batch_size       = 8,
    optimizer        = torch.optim.Adam,
    criterion        = nn.MSELoss,
    iterator_train__shuffle = True,
    callbacks        = [EarlyStopping(patience=10)],
    device           = "cuda" if torch.cuda.is_available() else "cpu",
    verbose          = 0,
)

param_grid = {
    "lr": [0.1, 0.01, 0.001],
    "module__hidden1": [16, 32, 64],
    "module__hidden2": [8, 16, 32],
    "batch_size": [8, 16, 32],
}

gs = GridSearchCV(
    estimator = net,
    param_grid= param_grid,
    cv        = 5,
    n_jobs    = -1,
    scoring   = "neg_mean_squared_error",
)
gs.fit(X_train, y_train)

print("Best params:", gs.best_params_)
print("Best CV MSE:", -gs.best_score_)

# 新模型在 Dag-1 上评估
best_net = gs.best_estimator_
with torch.no_grad():
    y_pred_dag1_new = best_net.predict(X_dag1)
acc_dag1_new = calculate_accuracy_within_threshold(y_dag1, y_pred_dag1_new, thr)
print(f"Accuracy on Dag-1 after GS = {acc_dag1_new:.2f}%")
```

```python
# =========================================================
# Section 7 ─ 生成三条序列（修正 .detach() / no_grad()）
# =========================================================
device = "cuda" if torch.cuda.is_available() else "cpu"

with torch.no_grad():  # 关闭 Autograd，提高推理效率
    # Bridge 段：旧模型预测
    bridge_pred_orig = (
        model_base(torch.from_numpy(X).to(device))
        .detach().cpu().numpy()
    )

    # Dag-1 段：旧模型预测
    dag1_pred_orig = (
        model_base(torch.from_numpy(X_dag1).to(device))
        .detach().cpu().numpy()
    )

# Dag-1 段：新模型预测（skorch 接口）
dag1_pred_new = best_net.predict(X_dag1)

# 真实值
bridge_true = y
dag1_true   = y_dag1

# 拼接
correction_offset = 38
join_index        = len(bridge_true)

# 原始预测（红线）
y_pred_selected = np.concatenate([bridge_pred_orig, dag1_pred_orig])

# 校正预测（绿线）
y_pred_selected_adjusted = np.concatenate([
    bridge_pred_orig,
    np.concatenate([dag1_pred_orig[:correction_offset],
                    dag1_pred_new[correction_offset:]])
])

# 真实值（蓝线）
y_selected_adjusted = np.concatenate([bridge_true, dag1_true])
```
